```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
#convert file into dataset
df = pd.read_csv('dataset-ecommerce_Farah.csv')
```

```python
df
```

#Data Manipulation

```python
#converting to date time
df["Order Date"] = pd.to_datetime(df["Order Date"])

#sorting fot data in chronological order
df = df.sort_values(by="Order Date")

#Menghilangkan $ dikolom sales
df['Sales'] = df['Sales'].str.replace("[^\d.]", "", regex=True)
df['Shipping Cost'] = df['Shipping Cost'].str.replace("[^\d.]", "", regex=True)

df
```

```python
df.info()
```

#Data Visualization

###Line Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Figure size in inches
plt.rcParams['figure.figsize'] =15, 7

#rotating x-axis values
plt.xticks(rotation = 'vertical')

#line plot
sns.lineplot(data=df, x='Order Date', y='Sales')

#labeling
plt.title(f"Sales Per Order", fontsize=18) #title
plt.xlabel("Date", fontsize=12) #x-axis label
plt.ylabel("Sales($)", fontsize=12) #y-axis label

#showing plot
plt.show()
```

```python

# Figure size in inches
plt.rcParams['figure.figsize'] =15, 7

#rotating x-axis values
plt.xticks(rotation = 'vertical')

#line plot
sns.lineplot(data=df, x='Order Date', y='Shipping Cost')

#labeling
plt.title(f"Shipping Cost Per Order", fontsize=18) #title
plt.xlabel("Date", fontsize=12) #x-axis label
plt.ylabel("Shipping Cost($)", fontsize=12) #y-axis label

#showing plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.lineplot(data=df, x='Product Category', y='Sales', ax=ax)


#labeling
plt.title(f"Sales Per Product Category", fontsize=18) #title
plt.xlabel("Product Category", fontsize=12) #x-axis label
plt.ylabel("Sales($)", fontsize=12) #y-axis label

#showing plot
plt.show
```

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.lineplot(data=df, x='Product Category', y='Sales', hue='Order Priority', ax=ax)


#labeling
plt.title(f"Sales per Product Category and Order Priority", fontsize=18) #title
plt.xlabel("Product Category", fontsize=12) #x-axis label
plt.ylabel("Sales($)", fontsize=12) #y-axis label

#showing plot
plt.show
```

### Pie Chart

```python
data = df['Ship Mode'].value_counts()
keys = df['Ship Mode'].unique()

#define seaborn color palette to use
palette_color = sns.color_palette('viridis')

#plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        autopct='%1.1f%%')

plt.title(f"Ship Mode", fontsize=18) #title

#displaying chart
plt.show()
```

**Months**

```python
a = df['Months'].value_counts()
b = df['Months'].unique()

#define seaborn color palette to use
palette_color = sns.color_palette('viridis')

#plotting data on chart
plt.pie(a, labels=b, colors=palette_color,
        autopct='%.0f%%')

plt.title(f"Months", fontsize=18) #title

#displaying chart
plt.show()
```

###Bar Chart

**Product Category**

```python
y = df['Product Category'].value_counts()
x = df['Product Category'].unique()

#Draw Bar Chart
sns.barplot(x=x, y=y, data = df, palette="Blues")

plt.show()
```
