#2.
Buatlah fungsi sum_even_numbers() yang melakukan penjumlahan hanya angka genap dalam
list. Fungsi sum_even_numbers() menerima input list of numbers dan mengeluarkan hasil
penjumlahan dari hanya angka genap yang ada pada list of numbers. (Score = 20)
   
    sum_even_numbers([4, 5, 7, 8])
    expected output: 12

```python
def sum_even(a):
	sum_ = 0
	for value in a:
		if value % 2 == 0:
			sum_ += value
	return sum_


print(sum_even([4, 5, 7, 8]))
```

Buatlah fungsi unique_list() yang menerima input list of numbers dan mengeluarkan list baru  yang berisi unique element. (Score = 30) 
Note: Dilarang menggunakan fungsi set(). 


    unique_list([4, 5, 5, 7, 7, 8]) 
    expected output: [4, 5, 7, 8]
    Note: Dilarang menggunakan fungsi set().  


```python
# dapatkan nilai unik dari list
# menggunakan numpy.unique
import numpy as np
 
# berfungsi untuk mendapatkan unique value
def unique(list1):
    x = np.array(list1)
    print(np.unique(x))
 
 
# driver code
unique_list = [4, 5, 5, 7, 7, 8]
print("the unique values from 1st list is")
unique(unique_list)
```

Buatlah sebuah class Circle dengan ketentuan sebagai berikut: (Score = 40) 

    a. attribute radius (jari-jari) 
    b. method area yang mengembalikan nilai luas lingkaran  
    c. method perimeter yang mengembalikan nilai keliling lingkaran 
    Note: gunakan math.pi untuk mendapatkan nilai pi 


```python
import math
class circle_compute():
   def __init__(x,jari2):
      x.jari2=jari2
   def area_kalkulasi(x):
      return math.pi*(x.jari2**2)
   def perimeter_kalkulasi(x):
      return 2*math.pi*x.jari2

# A.  Buat attribute radius (jari-jari = 7) 
hasil = int(input("Masukkan jari-jari lingkaran..."))
contoh = circle_compute(hasil)
print("Jari2 yang masuk adalah :")
print(hasil)

# B. Buat method area yang mengembalikan nilai luas lingkaran  
print("Luas lingkaran yang dihitung adalah ")
print(round(contoh.area_kalkulasi(),2))

# C. Buat method perimeter yang mengembalikan nilai keliling lingkaran 
print("Keliling lingkaran yang dihitung adalah :")
print(round(contoh.perimeter_kalkulasi(),2))
```
