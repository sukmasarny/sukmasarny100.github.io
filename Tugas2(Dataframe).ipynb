#1

Lakukan outer join pada table orders & table customers dengan menggunakan fungsi merge.
Tampilkan hanya kolom orderNumber, customerNumber, dan customerName. [Score:20]

```python


import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest
from google.colab import drive
drive.mount('/content/drive/') 

```

```python
df_orders = pd.read_csv('/content/drive/MyDrive/sukma/orders(1).csv')
df_orders.head()
```

```python
df_customers = pd.read_csv('/content/drive/MyDrive/sukma/customers(1).csv')
df_customers.head()
```

```python
df_orderdetails = pd.read_csv('/content/drive/MyDrive/sukma/orderdetails(1).csv')
df_orderdetails.head()
```

```python
df_data = df_orders.merge(df_customers, how='outer', on='customerNumber')
df_data.head()
```

```python
df_data[['orderNumber', 'customerNumber', 'customerName']]
```

Tampilkan top 7 customer dengan jumlah order terbanyak (bukan nominal terbanyak).
Tampilkan customerName beserta dengan jumlah ordernya. [Score: 30]

```python
df_data1 = df_data[['orderNumber', 'customerNumber', 'customerName']]\
.groupby(by=['customerName'])\
.agg({'orderNumber': ['sum']})


df_data1.head(7)
```

Tampilkan max totalAmount, mean totalAmount, count orderNumber untuk setiap negara.
Tampilkan terurut descending berdasarkan mean totalAmount. (Hint: merge table orders,
orderdetails, dan customers) [Score: 50]

```python
df_data[['orderNumber', 'country', 'creditLimit']]
```

```python
df_data2 = df_data[['orderNumber', 'country', 'totalAmount']]\
.groupby(by=['country'])\
.agg({'totalAmount': ['max,mean'],
     'orderNumber': ['count'] })\
     .sort_values([('orderNumber', 'count'), ('totalAmount', 'mean')], ascending=[False, False])
```
